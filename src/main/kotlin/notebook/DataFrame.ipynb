```kotlin
val personsDf = dataFrameOf(
    "Name" to listOf("Alice", "Bob", "Charlie"),
    "age" to listOf(15, 20, 100),
)
```

```kotlin
personsDf.describe()
```

```kotlin
personsDf.schema()
```

```kotlin
personsDf.select { Name }
```

```kotlin
personsDf.filter { age > 15 }
```

```kotlin
personsDf.rename { age }.into {
    val mean = it.data.mean()
    "age [mean = $mean]"
}
```

```kotlin
val renamed = personsDf.rename { age }.into { "Age" }
```

```kotlin
renamed
    .select { Age }
```
